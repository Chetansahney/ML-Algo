# *Guassian Naive Bayes*

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
#Generating the dataset which is normally distributed
x,y=make_classification(n_samples=1000,n_features=4,n_informative=2,n_redundant=2,random_state=42)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

n_informative: The features which contribute to class seperation, have their own existence.

n_redundant: Do not actually provide any much information, but is corelated with informative features.

```python
gnb=GaussianNB()
gnb.fit(x_train,y_train)
y_pred=gnb.predict(x_test)
a=accuracy_score(y_test,y_pred)
print("Accuracy Score:",a)
```

# *Multinomial Naive Bayes*

```python
from sklearn import datasets
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import accuracy_score
```

```python
data=fetch_20newsgroups(subset='all')
vectoriser=CountVectorizer()
x=vectoriser.fit_transform(data.data)
y=data.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

ðŸ“Œ What is the 20 Newsgroups dataset?

It is a collection of ~18,846 text documents grouped into 20 discussion topics (newsgroups) from early internet forums. Each document is basically a user-posted message from those groups.

```python
mnb=MultinomialNB()
mnb.fit(x_train,y_train)
y_pred=mnb.predict(x_test)
a=accuracy_score(y_test,y_pred)
print("Accuracy Score:",a)
```

# *Bernoulli Naive Bayes*

```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
```

```python
#Generating the dataset which is normally distributed
x,y=make_classification(n_samples=1000,n_features=4,n_informative=2,n_redundant=2,random_state=42)

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
bnb=BernoulliNB()
bnb.fit(x_train,y_train)
y_pred=bnb.predict(x_test)
a=accuracy_score(y_test,y_pred)
print("Accuracy Score:",a)
```
